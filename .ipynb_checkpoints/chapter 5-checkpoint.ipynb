```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
%matplotlib inline
```

```python
np.random.seed(1000) # 如果给一个seed,那么每次给出的随机数一样，如果不提前定义seed,那么每次都不一样
y = np.random.standard_normal(20)
x = range(len(y)) # 将x定义为0值19
plt.plot(x,y)
```

```python
plt.plot(y) #对于n纬数组，即使没有x，也能画出同样的图形
```

```python
plt.plot(y.cumsum())
plt.grid(True)
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.axis('tight') #调整坐标轴，以数据边界为限，调整坐标轴极限
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1,21) #设置x轴的左右边界
plt.ylim(np.min(y.cumsum()-1),np.max(y.cumsum()+1))#设置y轴的上下边界
```

```python
plt.figure(figsize=(10, 6)) #设定画面大小
plt.plot(y.cumsum(),'b',lw = 1.5) #蓝色线条，粗细1.5
plt.plot(y.cumsum(),'ro',) #黄色点
plt.grid(True) 
plt.xlim(-1,21) #设置x轴的左右边界
plt.ylim(np.min(y.cumsum()-1),np.max(y.cumsum()+1))#设置y轴的上下边界
plt.xlabel('time')
plt.ylabel('value')
plt.title('simple sample')
```

```python
np.random.seed(2000)
y =  np.random.standard_normal((20,2)).cumsum(axis=0) #y是2列20行随机数字。
plt.figure(figsize=(10,6))
plt.plot(y,lw = 1.5) #系统自动区分两条线，赋予不同颜色
plt.plot(y,'ro') #画顶点
plt.grid(True)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A sample of nuno')
```

```python
np.random.seed(2000)
y =  np.random.standard_normal((20,2)).cumsum(axis=0) #y是2列20行随机数字。
plt.figure(figsize=(10,6))
plt.plot(y[:,0],'b',lw = 1.5,label='1st') #系统自动区分两条线，赋予不同颜色
plt.plot(y[:,1],'y',lw = 1.5,label='2nd') #系统自动区分两条线，赋予不同颜色
plt.plot(y,'ro') #一次性2条线画顶点
plt.axis('tight') # plt.xlim...则可自定义区间
plt.grid(True)
plt.legend(loc=2) #加入图例，0-10代表不同位置
plt.xlabel('index')
plt.ylabel('value')
plt.title('A sample of nuno')
```

```python
np.random.seed(2000)
y =  np.random.standard_normal((20,2)).cumsum(axis=0) #y是2列20行随机数字。
y[:,0] = y[:,0]*100 #将第一组数据放大100倍，使两组数据具有不同的刻度，这样，用同一个y轴就无法画图了。 
plt.figure(figsize=(10,6))
plt.plot(y[:,0],'b',lw = 1.5,label='1st') #系统自动区分两条线，赋予不同颜色
plt.plot(y[:,1],'y',lw = 1.5,label='2nd') #系统自动区分两条线，赋予不同颜色
plt.plot(y,'ro') #一次性2条线画顶点
plt.axis('tight') # plt.xlim...则可自定义区间
plt.grid(True)
plt.legend(loc=2) #加入图例，0-10代表不同位置
plt.xlabel('index')
plt.ylabel('value')
plt.title('A sample of nuno')
```

此时有两种解决方案，画2个图，或者在同一个图中，使用左右2个y轴

```python
np.random.seed(2000)
y =  np.random.standard_normal((20,2)).cumsum(axis=0) 
y[:,0] = y[:,0]*100  
fig, ax1 = plt.subplots() #这是拆分的关键，这里有复数 "s"
plt.plot(y[:,0],'b',lw = 1.5,label='1st') 
plt.plot(y[:,0],'ro') 
plt.axis('tight') 
plt.grid(True)
plt.legend(loc=3) 
plt.xlabel('index')
plt.ylabel('value')
plt.title('A sample of nuno')
ax2 = ax1.twinx() #这也是拆分的关键
plt.plot(y[:,1],'g',lw = 1.5,label='2nd') 
plt.plot(y[:,1],'ro') 
plt.legend(loc=2) 
plt.ylabel('2nd value')
```

```python
plt.figure(figsize=(10,6))
plt.subplot(211) #并排或者竖直画几个图， 211表示2行图，1列图，这是第一个
plt.plot(y[:,0],'b',lw = 1.5,label='1st') 
plt.plot(y[:,0],'ro') 
plt.axis('tight') 
plt.grid(True)
plt.legend(loc=0) 
plt.xlabel('index')
plt.ylabel('value')
plt.title('A sample of nuno')
plt.subplot(212) #此后为第二个图
plt.plot(y[:,1],'b',lw = 1.5,label='2nd') 
plt.plot(y[:,1],'ro') 
plt.axis('tight') 
plt.grid(True)
plt.legend(loc=0) 
plt.xlabel('index')
plt.ylabel('value')
```

```python
plt.figure(figsize=(16,5))
plt.subplot(121) #横向排列图形
plt.plot(y[:,0],'b',lw = 1.5,label='1st') 
plt.plot(y[:,0],'ro') 
plt.axis('tight') 
plt.grid(True)
plt.legend(loc=0) 
plt.xlabel('index')
plt.ylabel('value')
plt.title('A sample of nuno')
plt.subplot(122) #第二个
plt.bar(np.arange(len(y)),y[:,1],width=0.5,color='g') #画柱状图,此处用np.arrange给出x。
plt.axis('tight') 
plt.grid(True)
plt.legend(loc=0) 
plt.xlabel('index')
plt.ylabel('value')
plt.title('Bar Chart')
```

画求积分的示意图

```python
from matplotlib.patches import Polygon
def f(x):
    return 0.5 * np.exp(x) + 1
a, b = 0.5, 1.5 # x取值的左右极限，用来计算积分
x = np.linspace(0,2) #获得线性向量
y = f(x)
fig, ax = plt.subplots(figsize=(7,5)) #同一图上分别画。 如果不分层，就画不出用来表示积分阴影的polygon
plt.plot(x,y,'b',linewidth=2)
plt.ylim(ymin=0)
Ix = np.linspace(a,b) 
Iy = f(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)] #勾勒出阴影多边形的边框，用加号连接。 list后是(); list叠加的方法+
poly = Polygon(verts, facecolor='0.7',edgecolor='0.5') #定义多边形
ax.add_patch(poly) #画出多边形。这里显示出分层画图的威力。

plt.text(0.5 *(a+b),1,r"$\int_a^b f(x)\mathrm{d}x$",horizontalalignment='center',fontsize=20) #如文字，位置，内容，对齐，字号
plt.figtext(0.9, 0.075, '$x$') #在x轴右端写x，写入内容用$$包括
plt.figtext(0.075, 0.9, '$f(x)$')#在y轴顶端端写入f(x)，写入内容用$$包括
ax.set_xticks((a,b)) #在x轴标出a,b两个点
ax.set_xticklabels(('$a$','$b$')) #写入x轴标签
ax.set_yticks([f(a),f(b)]) #确定y轴两个标签的位置
ax.set_yticklabels(('$f(a)$','$f(b)$'))#设定y轴标签
plt.grid(True) #自动在刚才确定的x,y轴的位置上画线

```

```python
import matplotlib.finance as mpf
start = (2016,3,18)
end = (2016,7,17)
quotes = mpf.quotes_historical_yahoo_ohlc('^GDAXI', start, end) #获取历史行情数据
fig, ax =  plt.subplots(figsize=(16,7)) #subplot一定要加s
fig.subplots_adjust(bottom=0.2) #将底端线上移0.2以便未来放标签
mpf.candlestick_ohlc(ax, quotes, width=0.6, colorup='r',colordown='b') #在ax这个轴的画面上画图
plt.grid(True) #加上这个以后就没有那么多的文字输出了
ax.xaxis_date() #将x轴设定为日期
ax.autoscale_view() #自动对齐两段，不会留空白
plt.setp(plt.gca().get_xticklabels(), rotation=30) #x轴标注旋转30度
```

实现股票图形+成交量

```python
quotes = np.array(mpf.quotes_historical_yahoo_ochl('YHOO',start,end)) #关键是要变成数组
fig, (ax1,ax2) = plt.subplots(2,figsize=(16,10)) #确定画2个分图层。之前的同图左右2个y轴。虽然也用subplots，但然后ax2 =  ax1.twins()，并没真的ax2
mpf.candlestick_ochl(ax1, quotes, colorup='r', colordown='b' ) #在ax2轴画蜡烛线
plt.grid(True)
ax1.set_title('Yahoo')
ax1.set_ylabel('Index Level')
ax1.xaxis_date()
plt.bar(quotes[:,0]-0.25, quotes[:, 5], width=0.5) #给定横轴参数，纵轴就是成交量，第一个变量是数组bin，表示每一个柱开始的位置
ax2.set_ylabel('volume')
ax2.grid(True)
ax2.autoscale_view()
plt.setp(plt.gca().get_xticklabels(), rotation=30)
```

```python
strike = np.linspace(50,150,24) #创建50到150的线性数组，line不含e
ttm = np.linspace(0.5,2.5,25) #创建time to matirity的数组
strike,ttm = np.meshgrid(strike,ttm) #将两个一维数组结合，构成二维数组
#本来，strike只是以一位数组，即50到150的一串数字。在经过meshgrid与ttm合并后，就变成二维数字，重复n行，每行都是这24个数字。
#n也是ttm中包含元素的数量。这样strike，ttm才能交叉成行二维数组。
iv = (strike - 100) ** 2 / (100 * strike) / ttm #假设implied volatility是这么算
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(16,10)) #figure返回的是图形窗口
ax = fig.gca(projection='3d')#返回当前axes的句柄。 figure--图形的窗口， axes--返回图形的轴
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2, cmap=plt.cm.coolwarm, linewidth=0.5, antialiased=True) #画“面”surf。
#命令是从轴出发，从ax.plot_surface，来画图。设置标签也是从轴出发
#rstride=2,  array row stride 步幅
#cstride=2, array column stride
#cmap=plt.cm.coolwarm a colormap for the surface patches
#linewidth=0.5, 
#antialiased=True
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.grid(True)
```

```python
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111, projection='3d') #从图形窗口，fig加入图形
ax.view_init(30, 60) #初始视角

ax.scatter(strike, ttm, iv, zdir='z', s=25,
           c='b', marker='^')

ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
```
