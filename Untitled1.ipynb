```python
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%pylab inline
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
```

```python
def f(r):
    out = np.sin(x)
    return out 
```

```python
figsize(10, 8)
plt.plot(x, f(x))
```

```python
reg = np.polyfit(x, f(x), deg=5)
```

```python
fy = np.polyval(reg, x)
```

```python
plt.plot(x, fy, label='np polyfit result @ deg equals 5')
xlabel('x values')
ylabel('fitting value of y')
title('np fitting')
```

```python
matrix = np.zeros((4, len(x)))
```

```python
matrix[0, :] = 1
matrix[1, :] = x
matrix[2, :] = x ** 2
matrix[3, :] = x ** 3
```

```python
reg = np.linalg.lstsq(matrix.T, f(x))
```

```python
reg_parmt = reg[0]
```

```python
lfy = np.dot(reg_parmt, matrix)
```

```python
plt.plot(x, lfy)
```
